Para isso, basta passar o caminho da imagem a ser analisada e o programa irá retornar a imagem com uma *bouding-box* ao redor do objeto e sua respectiva categoria.

```
import cv2
import cvlib as cv
from cvlib.object_detection import draw_bbox
import matplotlib.pyplot as plt
```

## Leitura da imagem
Para ler a imagem basta passar o caminho do arquivo de foto a ser analisado. Em seguida, será demonstrado uma visualização da imagem.

```
caminho_imagem = 'drive/My Drive/Imagens/000000231822.jpg'
img = cv2.imread(caminho_imagem)
plt.imshow(img[...,::-1])
plt.show()
```

##Detecção dos objetos
Para detectar as objetos basta utilizarmos o método *detect_common_objects*, passando o objeto imagem, da biblioteca cv2. Esse método retorna as coordenadas da *bounding-box* do objeto, a categoria atribuída e a confiança atribuída. 
Dessa forma, basta apenas agora criar as *bouding-box* na imagem original, para isso, podemos usar o método *draw_bbox*, passando os parâmetros retornado pelo método *detect_common_objects*. 
Feito isso, podemos passar o objeto recebido para visualização com o matplotlib.

```
bbox, categorias, confiancas = cv.detect_common_objects(img)

nova_imagem = draw_bbox(img, bbox, categorias, confiancas)

plt.imshow(nova_imagem[...,::-1])
plt.show()
```
